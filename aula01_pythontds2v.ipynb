```python
# Importando a biblioteca pandas

import pandas as pd
```

```python
# Importando o data frame salaries

df = pd.read_csv("https://raw.githubusercontent.com/fvcarrijo/python/refs/heads/main/salaries.csv")
```

```python
df.columns
```

```python
df["experience_level"].value_counts()
```

```python
experience_level_mapping = {
    'SE': 'Senior',
    'MI': 'Pleno',
    'EN': 'Júnior',
    'EX': 'Executivo'
}
df['experience_level'] = df['experience_level'].replace(experience_level_mapping)
print(df['experience_level'].value_counts())
```

```python
df.head()
```

```python
df["employment_type"].value_counts()
```

```python
employment_type_mapping = {
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Meio Período',
    'FL': 'Freelancer'
}
df['employment_type'] = df['employment_type'].replace(employment_type_mapping)
print(df['employment_type'].value_counts())
```

```python
df.head()
```

```python
df["company_size"].value_counts()
```

```python
column_mapping = {
    'work_year': 'ano_trabalho',
    'experience_level': 'nivel_experiencia',
    'employment_type': 'tipo_emprego',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda_salario',
    'salary_in_usd': 'salario_em_dolar',
    'employee_residence': 'endereco',
    'remote_ratio': 'taxa',
    'company_location': 'filial',
    'company_size': 'porte_empresa'
}
df = df.rename(columns=column_mapping)
print(df.columns)
```
