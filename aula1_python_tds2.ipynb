```python
# Importando a biblioteca pandas

import pandas as pd
```

```python
# Busca o arquivo do data frame salaries

df = pd.read_csv("https://raw.githubusercontent.com/fvcarrijo/python/refs/heads/main/salaries.csv")
```

```python
# Linhas e colunas do data frame em um formato mais amigável

linhas,colunas = df.shape[0],df.shape[1]
print("linhas: ", linhas)
print("colunas: ", colunas)
```

```python
# Colunas do data frame

df.columns
```

```python
# Sumarizando a coluna experience_level do data frame

df["experience_level"].value_counts()
```

```python
# Resumindo os dados da coluna company_size

df["company_size"].value_counts()
```

```python
# Otimizando a sumarização dos dados da coluna company_size
tamanho_empresa = {
    'S': 'Pequena',
    'M': 'Média',
    'L': 'Grande'
}
df['company_size'] = df['company_size'].replace(tamanho_empresa)
display(df['company_size'].value_counts())
```

```python
# Alterando o nome da coluna company_size para tamanho_empresa

df = df.rename(columns={'company_size': 'tamanho_empresa'})
df.head()
```
